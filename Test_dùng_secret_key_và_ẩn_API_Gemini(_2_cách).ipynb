```python
# Install the required library (if not already installed)
!pip install --upgrade google-generativeai

import os
import google.generativeai as genai
from google.colab import userdata

# 1. Retrieve API Key from Colab Secrets
api_key = userdata.get('GOOGLE_API_KEY')  # Replace 'GEMINI_API_KEY' with your secret name

# 2. Handle Missing API Key
if not api_key:
    raise ValueError("API Key not found in Colab Secrets. Please create a secret named 'GEMINI_API_KEY'.")

# 3. Configure Google Generative AI Client
genai.configure(api_key=api_key)

# 4. Load the Gemini Model
try:
    model = genai.GenerativeModel(model_name="gemini-pro")  # Use "gemini-pro" if "gemini-2.0-flash" is unavailable
except Exception as e:
    print("Error loading the model:", e)
    raise

# 5. Define your prompt
question = """Write a 1000 word blog post about Hướng dẫn sử dụng nhanh n8n? Trả lời bằng văn bản thuần túy.
Không sử dụng dấu gạch đầu dòng (-), dấu phân cách (---), hoặc bất kỳ định dạng Markdown nào.
Trả lời bằng tiếng Việt."""

# 6. Generate a response using the loaded model
try:
    response = model.generate_content(question)  # Use the question directly as the prompt

    # Post-process the response to ensure plain text formatting
    formatted_response = response.text.strip()  # Remove leading/trailing whitespace
    formatted_response = formatted_response.replace("*", "")  # Remove any stray * characters
    formatted_response = formatted_response.replace("---", "")  # Remove any stray ---
    formatted_response = formatted_response.replace("**", "")  # Remove bold formatting

    # Print the formatted response
    print("Gemini API's response:\n")
    print(formatted_response)
except Exception as e:
    print("Error occurred while generating content:", e)
```
