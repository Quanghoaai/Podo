```python
import google.generativeai as genai

# Giả định đã đặt API key (ví dụ: sử dụng biến môi trường)
import os
# genai.configure(api_key=os.environ["GOOGLE_API_KEY"]) # Cách bảo mật: Lấy từ biến môi trường

# Thay thế bằng API key của bạn (nếu không dùng biến môi trường)
# KHÔNG NÊN ĐƯA API KEY TRỰC TIẾP VÀO CODE
# Dùng tạm thời (không khuyến khích):
GOOGLE_API_KEY = "AIzaSyAwm7NOQXENkOddEcvqACCPUX-McF0sSsk"  # Thay YOUR_API_KEY bằng API key thật
genai.configure(api_key=GOOGLE_API_KEY)



# Các prompts (ví dụ)
prompts = {
    1: "Bạn là một trợ lý đầu tư. Hãy giải thích các bước...",
    2: "Bạn là một chuyên gia tư vấn đầu tư. Dựa trên thông tin...",
    3: "Bạn là một chuyên gia phân tích thị trường. Dựa trên...",
    4: "Bạn là một cố vấn đầu tư, Hãy đề xuất...",
    5: "Bạn là một hệ thống hỗ trợ giao dịch...",
    6: "Bạn là một hệ thống quản lý danh mục đầu tư...", # Cẩn thận!
    7: "Bạn là một trợ lý đầu tư. Hãy trả lời câu hỏi sau...",
    8: "Bạn là một chuyên gia quản lý rủi ro. Hãy phân tích...",
    9: "Bạn là một chuyên gia tối ưu hóa danh mục đầu tư...",
    10: "Bạn là một hệ thống báo cáo. Hãy tạo báo cáo..."
}

def classify_request(request):
  # Phân loại yêu cầu (ví dụ đơn giản)
    request = request.lower()
    if "tạo danh mục" in request:
        return 2  # Chọn loại danh mục
    elif "loại tài sản" in request:
        return 3  # Chọn tài sản
    elif "chiến lược" in request:
        return 4  # Chọn chiến lược
    elif "điều kiện" in request:
        return 5
    # ... (thêm các loại yêu cầu khác) ...
    else:
        return 7 #Hỏi đáp

def get_response(prompt, data=None):
    model = genai.GenerativeModel('gemini-pro') # Chọn model
    chat = model.start_chat() #Nếu dùng chế độ liên tục

    if data:
        prompt += f"\n\nDữ liệu/Ngữ cảnh:\n{data}" # Nối thêm data vào prompt
        # print(prompt) # Dùng để kiểm tra prompt cuối cùng
    response = chat.send_message(prompt)
    return response.text

def post_process(response):
  # Xử lý hậu kỳ (nếu cần)
    return response

def handle_request(request, data=None):
    prompt_id = classify_request(request)
    prompt = prompts[prompt_id]
    response = get_response(prompt, data=data)
    final_response = post_process(response)
    return final_response
# Ví dụ
user_request = "Tôi muốn tạo một danh mục đầu tư tăng trưởng tại thị trường Việt Nam"
user_profile = """
Tuổi: 30
Mục tiêu: Tăng trưởng vốn dài hạn
Mức độ chấp nhận rủi ro: Trung bình
"""

response = handle_request(user_request, data=user_profile)
print(response)

user_request = "Nên chọn những loại tài sản nào?"
response = handle_request(user_request) # Không truyền data cũng được.
print(response)
```
